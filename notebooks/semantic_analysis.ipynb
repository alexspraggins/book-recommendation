```python
# # LOADS Text Classification MODEL FROM DRIVE (Device: GPU=0, CPU=-1) (Check for GPU, !nvidia-smi)

from transformers import pipeline

model_path = "/content/drive/MyDrive/Colab Notebooks/Projects/Book Recommendation/venv_cache/text_classification"

classifier = pipeline(
    "text-classification",
    model=model_path,
    tokenizer=model_path,
    top_k=None,
    device=0
)
```

```python
import pandas as pd
import numpy as np

books = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Projects/Book Recommendation/books/books_simple_prediction.csv")
```

```python
books
```

```python
emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise"]
```

```python
from typing import Any, List
from tqdm.notebook import tqdm

def calculate_max_emotion_scores(predictions: Any, emotion_labels: List[str]) -> dict[str: float]:
  # Given sentences, store emotion predicitons for each sentence
  per_emotion_scores = {label:[] for label in emotion_labels}

  for prediction in predictions: # Each sentence of description
    # Sort by emotion so index & label match
    sorted_emotions = sorted(prediction, key=lambda x: x["label"])

    # Add emotion score to emotions for all sentences
    for index, label in enumerate(emotion_labels):
      per_emotion_scores[label].append(sorted_emotions[index]["score"])

  # Get the maximum for each emotion
  return {label: np.max(scores) for label, scores in per_emotion_scores.items()}

def generate_emotions_df(books: pd.DataFrame, classifier, emotion_labels) -> pd.DataFrame:
  isbn = []
  emotion_scores = {label: [] for label in emotion_labels}

  for i in tqdm(range(len(books))):
    isbn.append(books.iloc[i]["isbn13"])
    sentences = books.iloc[i]["description"].split(".")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions, emotion_labels)

    for label in emotion_labels:
      emotion_scores[label].append(max_scores[label])

  emotions_df = pd.DataFrame(emotion_scores)
  emotions_df["isbn13"] = isbn

  return emotions_df

def merge_emotions_with_books(books, emotions_df) -> pd.DataFrame:
  books = pd.merge(books, emotions_df, on="isbn13", how="left")
  return books
```

```python
emotions_df = generate_emotions_df(books, classifier, emotion_labels)
```

```python
emotions_df
```

```python
books = merge_emotions_with_books(books, emotions_df)
```
